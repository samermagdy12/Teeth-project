```python
import os
import shutil

# Source directories
train_dir = r"D:\Teeth DataSet\Training"
val_dir = r"D:\Teeth DataSet\Validation"
test_dir = r"D:\Teeth DataSet\Testing"

# Destination merged directory
merged_dir = r"D:\Teeth DataSet\Merged"

# List of all dataset splits
source_dirs = [train_dir, val_dir, test_dir]

# Create merged root folder
os.makedirs(merged_dir, exist_ok=True)

# Loop through each split (train/val/test)
for src_root in source_dirs:
    for class_name in os.listdir(src_root):
        src_class_path = os.path.join(src_root, class_name)

        # Skip if not a directory
        if not os.path.isdir(src_class_path):
            continue

        # Create class folder in merged directory
        dst_class_path = os.path.join(merged_dir, class_name)
        os.makedirs(dst_class_path, exist_ok=True)

        # Copy each image
        for file_name in os.listdir(src_class_path):
            src_file = os.path.join(src_class_path, file_name)
            dst_file = os.path.join(dst_class_path, file_name)

            # Rename if duplicate filename exists
            if os.path.exists(dst_file):
                name, ext = os.path.splitext(file_name)
                counter = 1
                while True:
                    new_name = f"{name}_{counter}{ext}"
                    new_dst = os.path.join(dst_class_path, new_name)
                    if not os.path.exists(new_dst):
                        dst_file = new_dst
                        break
                    counter += 1

            shutil.copy2(src_file, dst_file)

print("All data merged successfully into 'Merged' folder!")
```
